```python
# impor libraries for the project
import numpy as np                       
import pandas as pd                                  
import matplotlib.pyplot as plt
import seaborn as sns                       
```

```python
# import penguine data as a data source for the project
penguine = pd.read_csv('penguins.csv')

# check the first 5 rows of the data
penguine.head()
```

```python
# There is some nan values in the data, so we will drop them
penguine.info()
```

```python
# check the sum of na data in the dataframe
penguine.isna().sum()
```

```python
# drop na associated with the sex column of the dataframe
penguine_clean = penguine.dropna(subset='sex')
penguine_clean.isna().sum()
```

```python
# understand datatypes within the dataframe
penguine_clean.info()
```

```python
penguine_categorical = penguine_clean.select_dtypes(include='object')
pen_cat = penguine_clean[['species', 'island', 'sex']]
```

```python
# check unique variables withing each column of the dataframe

for column in penguine_categorical.columns:
    print(f'{column}: {penguine_clean[column].unique()}')
```

```python
# pair plot of the cleaned data colored by species
# plt.figure(figsize=(10, 10))
# sns.set(style='whitegrid')
sns.pairplot(penguine_clean, hue='species', palette='Set2')
plt.show()
```

```python
sns.pairplot(penguine_clean, hue='sex')
plt.show()
```

```python
sns.pairplot(penguine_clean, hue='island')
plt.show()
```

```python
# box plot with species and island as the x and y axis respectively
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=penguine_clean, palette='Set2')
sns.stripplot(x='species', y='flipper_length_mm', data=penguine_clean, color='black',jitter=True)
plt.title('Box plot of species and island')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
# box plot with species and island as the x and y axis respectively
sns.boxplot(x='species', y='bill_length_mm', data=penguine_clean, palette='Set2', ax=ax1)
sns.stripplot(x='species', y='bill_length_mm', data=penguine_clean, color='black', jitter=True, ax=ax1)
ax1.set_title('Box plot of species and bill length')

# box plot with species and island as the x and y axis respectively
sns.boxplot(x='species', y='bill_depth_mm', data=penguine_clean, palette='Set2', ax=ax2)
sns.stripplot(x='species', y='bill_depth_mm', data=penguine_clean, color='black', jitter=True, ax=ax2)
ax2.set_title('Box plot of species and bill depth')
plt.show()
```
